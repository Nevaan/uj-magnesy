```python
filename = "accept-1.5.csv"
vals = [
    "0.0",
    "1.0",
    "1.5",
    "2.0",
    "3.0",
    "5.0"
]
import matplotlib.pyplot as plt
import pandas as pd
import csv
```

```python
def plotAccept(filename):
    headers = ['Step','AcceptRatio']
    df = pd.read_csv(filename,names=headers)
    x = df['Step']
    y = df['AcceptRatio']
    plt.plot(x,y)
    plt.show()
```

```python
def plotMagnet(filename):
    headers = ['Step','ChangedMagnets']
    df = pd.read_csv(filename,names=headers)
    x = df['Step']
    y = df['ChangedMagnets']
    plt.plot(x,y)
    plt.show()
```

```python
for i in vals:
    print(f'Values for TkB: {i}')
    plotAccept(f'accept-{i}.csv')
    plotMagnet(f'magnet-{i}.csv')
```
